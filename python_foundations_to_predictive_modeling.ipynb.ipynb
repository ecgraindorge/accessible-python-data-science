## 2.3 Dictionaries

A **dictionary** stores information using **key-value pairs**.

Think of a dictionary as a set of labeled containers:

| Key | Value |
|---|---|
| `"name"` | `"Alex"` |
| `"program"` | `"Applied Data Science"` |

Dictionaries are useful when each value needs a clear label. In data science, dictionaries often appear when working with:

````python
student = {
    "name": "Alex",
    "program": "Applied Data Science",
    "quiz_score": 87.5,
    "passed": True
}

print(student)
print(type(student))
````

## 2.4 Accessing Values in a Dictionary

To access a value, use its key inside square brackets.

The pattern is:

```python
dictionary_name["key_name"]
```

Keys are case-sensitive. For example, `"name"` and `"Name"` would be treated as different keys.

````python
print(student["name"])
print(student["program"])
print(student["quiz_score"])
````

## 2.5 Using `.get()` to Avoid Errors

If you try to access a key that does not exist, Python gives a `KeyError`.

The `.get()` method is safer when you are not sure whether a key exists.

The pattern is:

```python
dictionary_name.get("key_name", default_value)
```

The default value is what Python returns if the key is missing.

````python
# This key exists, so Python returns the value.
print(student.get("quiz_score"))

# This key does not exist, so Python returns None by default.
print(student.get("grade_level"))

# We can also provide our own default message.
print(student.get("grade_level", "Not listed"))

# This would cause a KeyError because the key does not exist:
# print(student["grade_level"])
````

## 2.6 Adding and Updating Dictionary Values

Dictionaries can be changed after they are created.

You can:

- add a new key-value pair
- update the value for an existing key
- create a new value based on existing values

````python
# Add a new key-value pair.
student["hours_studied"] = 6

# Update an existing value.
student["quiz_score"] = 91.0

# Create a new value using existing information.
student["needs_follow_up"] = student["quiz_score"] < 70

print(student)
````

## 2.7 Removing Values from a Dictionary

The `.pop()` method removes a key-value pair and returns the value that was removed.

This can be useful when cleaning data or removing information that is no longer needed.

````python
removed_value = student.pop("needs_follow_up")

print("Removed value:", removed_value)
print(student)
````

## 2.8 Looking at Keys, Values, and Items

Dictionaries have three helpful methods:

| Method | What it gives you |
|---|---|
| `.keys()` | all keys |
| `.values()` | all values |
| `.items()` | all key-value pairs |

The `.items()` method is especially useful when looping through dictionaries.

````python
print("Keys:")
for key in student.keys():
    print(key)

print("\nValues:")
for value in student.values():
    print(value)

print("\nKey-value pairs:")
for key, value in student.items():
    print(f"{key}: {value}")
````

# Part 8: Creating a Practice Dataset

For the rest of the notebook, we will use a **synthetic student success dataset**.

Synthetic means the data is made up for learning purposes.  
It is not real student data.

The dataset includes variables such as:

- study hours
- attendance rate
- assignment completion rate
- number of late submissions
- prior coding experience
- job hours
- whether the student passed the course

Our goal later will be to predict whether a student passed the course.

````python
# Set a random seed so everyone gets the same dataset
np.random.seed(42)

n = 250

student_id = np.arange(1, n + 1)
study_hours_per_week = np.random.normal(loc=6, scale=2.5, size=n).clip(0, 15)
attendance_rate = np.random.normal(loc=84, scale=12, size=n).clip(40, 100)
assignment_completion_rate = np.random.normal(loc=82, scale=15, size=n).clip(20, 100)
late_submissions = np.random.poisson(lam=1.5, size=n).clip(0, 8)
job_hours_per_week = np.random.normal(loc=20, scale=10, size=n).clip(0, 50)
sleep_hours = np.random.normal(loc=6.8, scale=1.1, size=n).clip(3, 10)

prior_coding_experience = np.random.choice(["None", "Some", "Strong"], size=n, p=[0.45, 0.40, 0.15])
internet_access = np.random.choice(["Reliable", "Unreliable"], size=n, p=[0.82, 0.18])

# Create a probability of passing based on several features.
# This is only for creating a realistic synthetic target variable.
experience_bonus = np.where(prior_coding_experience == "Strong", 0.8,
                   np.where(prior_coding_experience == "Some", 0.35, 0))
internet_penalty = np.where(internet_access == "Unreliable", -0.5, 0)

logit = (
    -5
    + 0.35 * study_hours_per_week
    + 0.045 * attendance_rate
    + 0.035 * assignment_completion_rate
    - 0.35 * late_submissions
    - 0.025 * job_hours_per_week
    + 0.20 * sleep_hours
    + experience_bonus
    + internet_penalty
)

prob_passed = 1 / (1 + np.exp(-logit))
passed_course = np.random.binomial(1, prob_passed)

df = pd.DataFrame({
    "Student ID": student_id,
    "Study Hours Per Week": study_hours_per_week.round(1),
    "Attendance Rate": attendance_rate.round(1),
    "Assignment Completion Rate": assignment_completion_rate.round(1),
    "Late Submissions": late_submissions,
    "Job Hours Per Week": job_hours_per_week.round(1),
    "Sleep Hours": sleep_hours.round(1),
    "Prior Coding Experience": prior_coding_experience,
    "Internet Access": internet_access,
    "Passed Course": passed_course
})

# Add a few missing values to practice data cleaning
missing_indices = np.random.choice(df.index, size=20, replace=False)
df.loc[missing_indices[:5], "Study Hours Per Week"] = np.nan
df.loc[missing_indices[5:10], "Attendance Rate"] = np.nan
df.loc[missing_indices[10:15], "Prior Coding Experience"] = np.nan
df.loc[missing_indices[15:20], "Internet Access"] = np.nan

# Add a duplicate row to practice duplicate detection
df = pd.concat([df, df.iloc[[0]]], ignore_index=True)

# Save a CSV version in case students want to practice loading a file
df.to_csv("student_success_synthetic.csv", index=False)

print("Synthetic dataset created and saved as student_success_synthetic.csv")
df.head()
````

````python
print("Rows before removing duplicates:", df_clean.shape[0])

df_clean = df_clean.drop_duplicates()

print("Rows after removing duplicates:", df_clean.shape[0])
````

## 10.4 Handle Missing Numeric Values

For numeric columns, a common simple approach is to fill missing values with the median.

The median is often better than the mean when there may be unusual high or low values.

````python
numeric_columns = [
    "study_hours_per_week",
    "attendance_rate",
    "assignment_completion_rate",
    "late_submissions",
    "job_hours_per_week",
    "sleep_hours"
]

for col in numeric_columns:
    median_value = df_clean[col].median()
    df_clean[col] = df_clean[col].fillna(median_value)
    print(f"Filled missing values in {col} with median: {median_value}")
````

## 10.5 Handle Missing Categorical Values

For categorical columns, one simple approach is to fill missing values with `"Unknown"`.

This preserves the fact that the value was missing.

````python
categorical_columns = [
    "prior_coding_experience",
    "internet_access"
]

for col in categorical_columns:
    df_clean[col] = df_clean[col].fillna("Unknown")

df_clean.isna().sum()
````

## 10.6 Create a New Feature

Feature engineering means creating new variables from existing variables.

For example, we can create an `engagement_score` using study hours, attendance, and assignment completion.

````python
df_clean["engagement_score"] = (
    df_clean["study_hours_per_week"] * 2
    + df_clean["attendance_rate"] * 0.5
    + df_clean["assignment_completion_rate"] * 0.5
    - df_clean["late_submissions"] * 3
)

df_clean[["study_hours_per_week", "attendance_rate", "assignment_completion_rate", "late_submissions", "engagement_score"]].head()
````

# Part 12: Introduction to Predictive Modeling

Important vocabulary:

| Term | Meaning |

| Target | The outcome we want to predict |
| Training data | Data used to teach the model |
| Testing data | Data used to evaluate the model |
| Classification | Predicting a category |
| Regression | Predicting a number |

Because `passed_course` is either 0 or 1, this is a **classification** problem.

## 12.1 Choose Features and Target

We will avoid using `student_id` because it is only an identifier.  
An ID number should not help predict the outcome.

````python
features = [
    "study_hours_per_week",
    "attendance_rate",
    "assignment_completion_rate",
    "late_submissions",
    "job_hours_per_week",
    "sleep_hours",
    "prior_coding_experience",
    "internet_access",
    "engagement_score",
    "attendance_category",
    "study_level"
]

target = "passed_course"

X = df_clean[features]
y = df_clean[target]

X.head()
````

## 12.2 Split into Training and Testing Data

We do not evaluate a model on the same data it learned from.

Instead, we split the dataset into:

- a training set
- a testing set

The model learns from the training set.  
Then we evaluate it on the testing set.

````python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

print("Training rows:", X_train.shape[0])
print("Testing rows:", X_test.shape[0])
````

## 12.3 Create a Baseline Model

A baseline is a simple comparison point.

For classification, a simple baseline is:  
**Always predict the most common class.**

If our machine learning model cannot beat this baseline, it may not be useful.

````python
most_common_class = y_train.mode()[0]
baseline_predictions = np.repeat(most_common_class, len(y_test))

baseline_accuracy = accuracy_score(y_test, baseline_predictions)

print("Most common class:", most_common_class)
print("Baseline accuracy:", baseline_accuracy)
````

# Part 13: Preparing Data for Machine Learning

Machine learning models usually need numeric input.

But our dataset has both:

- numeric columns
- categorical columns

We will use a preprocessing pipeline:

- Numeric columns will be standardized.
- Categorical columns will be one-hot encoded.

One-hot encoding turns categories into 0/1 columns.

````python
numeric_features = [
    "study_hours_per_week",
    "attendance_rate",
    "assignment_completion_rate",
    "late_submissions",
    "job_hours_per_week",
    "sleep_hours",
    "engagement_score"
]

categorical_features = [
    "prior_coding_experience",
    "internet_access",
    "attendance_category",
    "study_level"
]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ]
)

print("Preprocessor created.")
````

# Part 14: Build a Logistic Regression Model

Logistic regression is a common beginner-friendly model for binary classification.

Despite the name "regression," logistic regression is used for classification when the target has two classes.

````python
model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("classifier", LogisticRegression(max_iter=1000))
    ]
)

model.fit(X_train, y_train)

print("Model trained successfully!")
````

## 14.1 Make Predictions

````python
y_pred = model.predict(X_test)

print(y_pred[:10])
````

# Part 15: Evaluate the Model

Accuracy tells us the percentage of correct predictions.

But accuracy is not always enough.

We will also look at:

- Precision
- Recall
- Confusion matrix

For this example:

- Precision asks: Of the students predicted to pass, how many actually passed?
- Recall asks: Of the students who actually passed, how many did the model correctly identify?

````python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Accuracy:", round(accuracy, 3))
print("Precision:", round(precision, 3))
print("Recall:", round(recall, 3))
````

````python
print(classification_report(y_test, y_pred))
````

## 15.1 Confusion Matrix

A confusion matrix compares actual outcomes to predicted outcomes.

````python
cm = confusion_matrix(y_test, y_pred)

cm_df = pd.DataFrame(
    cm,
    index=["Actual Did Not Pass", "Actual Passed"],
    columns=["Predicted Did Not Pass", "Predicted Passed"]
)

cm_df
````

## 15.2 Visualizing the Confusion Matrix

````python
plt.imshow(cm)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks([0, 1], ["Did Not Pass", "Passed"])
plt.yticks([0, 1], ["Did Not Pass", "Passed"])

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.show()
````
